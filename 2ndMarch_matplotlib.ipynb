### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


ans. A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title("this is example of scatter plot")
```

### Q3: Why is the subplot() function used?

ans. Matplotlib’spyplot API has a convenience function called subplots() which acts as a utility wrapper and helps in creating common layouts of subplots, including the enclosing figure object, in a single call.

 Draw four line plots using the subplot() function.

Use the following data:

```python
import numpy as np
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(4, 1, 1)
plt.plot(x,y)
plt.title("1st plot")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(4, 1, 2)
plt.plot(x,y)
plt.title("2nd plot")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(4, 1, 3)
plt.plot(x,y)
plt.title("3rd plot")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(4, 1, 4)
plt.plot(x,y)
plt.title("4th plot")
```

### Q4: What is a bar plot? Why is it used? 

ans. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

### Q5: What is a box plot? Why is it used?

ans. A box plot is defined as a graphical method of displaying variation in a set of data. In most cases, a histogram analysis provides a sufficient display, but a box plot can provide additional detail while allowing multiple sets of data to be displayed in the same graph.

 Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
data_set = [box1,box2]
plt.boxplot(data_set)
```
